```python
#cargar los tags
import pandas as pd
import numpy as np
import copy
df = pd.read_csv('tags_deforestation.csv')
```

# New Section

```python
import os
import shutil
carpeta_fuente = '/content/train-jpg-e/'
carpeta_destino = '/content/dataset_augmented/'
total_entrenamiento = 30384 
df_train_tags = copy.deepcopy(df.iloc[:total_entrenamiento])  
```

```python
import cv2
import shutil
```

```python
tags = []
for i in range(df_train_tags.shape[0]):
  img = cv2.imread(carpeta_fuente+'train_'+str(i)+'_e.jpg')
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  imgvf = cv2.flip(img,0)
  imghf = cv2.flip(img,1)
  imgdf = cv2.flip(imgvf,1)
  cv2.imwrite(carpeta_destino+'train_'+str(i)+'_e.jpg',img)
  cv2.imwrite(carpeta_destino+'train_'+str(i)+'_evf.jpg',imgvf)
  cv2.imwrite(carpeta_destino+'train_'+str(i)+'_ehf.jpg',imghf)
  cv2.imwrite(carpeta_destino+'train_'+str(i)+'_edf.jpg',imgdf)
  tags.append(df_train_tags.deforestation[i])
  tags.append(df_train_tags.deforestation[i])
  tags.append(df_train_tags.deforestation[i])
  tags.append(df_train_tags.deforestation[i])
```

```python
df_train = pd.DataFrame(tags,columns=['deforestation'])
```

```python
df_train.to_csv('train_tags.csv')
```
